#Matplotlib - Gráfico de linha - Parte 1

```python
#Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X
X = np.linspace(0,5,11)
```

```python
# Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
# Gráfico de Y em função de X
plt.plot(X,Y)
plt.show()
```

```python
#Inserindo nomes nos eixos
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Inserindo título no gráfico
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois gráficos na mesma imagem
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
#Inserindo legenda
plt.plot(X,Y,color='black',label='Y(X)')
plt.plot(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
