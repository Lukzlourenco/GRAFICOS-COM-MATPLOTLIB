**2D_Grid**

**Manipulando a grade (malha, grid)** 


```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Criando Váriavel X
X = np.linspace(0,5,11)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#Criando Grid
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle ='-' , which = 'major', color = 'grey')
plt.legend()
plt.show()
```

```python
#Alterando a cor de fundo

plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle ='--' , which = 'major', color = 'grey')
plt.legend()
plt.show()
```

**Gráfico de Histograma**

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Criando variável X com números aleatórios
X= np.random.rand(2000)
```

```python
plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
#Alterando número de bins

plt.hist(X, bins=10)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=10')
plt.show()
```

```python
plt.hist(X, bins=20)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X, bins=30)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=30')
plt.show()
```

```python
#Alterando cores

plt.hist(X, bins=50, color='red')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X, bins=50, color='red', edgecolor='blue')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#Múltiplos histogramas na mesma imagem

X1 = np.random.normal(0,1, size = 800)
X2 = np.random.normal(3,1, size = 800)
X3 = np.random.normal(5,1, size = 800)
```

```python
plt.hist(X1, bins=10, color='red')
plt.hist(X2, bins=10, color='blue')
plt.hist(X3, bins=10, color='green')

plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
#Distribuição cumulativa

X = np.random.rand(2000)
```

```python
plt.hist(X, bins=20, color='red', edgecolor='blue', cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```

**Matplotlib - Legenda**

```python
#Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Criando variável X
X = np.linspace(0,5,11)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
# Colocando equações na legenda
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=E^{(X/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
#Adicionando posição da legenda 
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend(loc='lower right')
plt.show()
```

```python
#Criando novas variáveis X, Y e Z 

X = np.linspace(0,2.0*np.pi,361)
Y = np.cos(X/2)
Z = np.sin(X/4)
```

```python
plt.plot(X,Y,label='$\cos(X/2)$',color='red')
plt.plot(X,Z,label='$\sin(X/4)$',color='green',linestyle='--')
plt.legend()
```

**Matplotlib - Subplot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,5,11)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.subplot(2,1,1)
plt.plot(X,Y)
plt.subplot(2,1,2)
plt.plot(X,Z)
plt.show()
```

```python
#Nomeando eixos de cada gráfico
plt.subplot(2,1,1)
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.subplot(2,1,2)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Z')
plt.tight_layout()
plt.show()
```

```python
#Usando notação de objeto
plt.subplot(1,2,1)
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.subplot(1,2,2)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Z')
plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(X,Y)
ax[0].set_xlabel('Eixo X')
ax[0].set_ylabel('Eixo Y')
ax[1].plot(X,Z)
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Z')
plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1)
ax[0].plot(X,Y)
ax[0].set_xlabel('Eixo X')
ax[0].set_ylabel('Eixo Y')
ax[1].plot(X,Z)
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Z')
plt.tight_layout()
```

```python
#Subplot 2X2
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
W = np.power(X,3)
L = np.power(X,0.5)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)
ax[0][0].plot(X,Y)
ax[0][0].set_xlabel('Eixo X')
ax[0][0].set_ylabel('Eixo Y')
ax[0][1].plot(X,Z)
ax[0][1].set_xlabel('Eixo X')
ax[0][1].set_ylabel('Eixo Z')

ax[1][0].plot(X,W)
ax[1][0].set_xlabel('Eixo X')
ax[1][0].set_ylabel('Eixo W')
ax[1][1].plot(X,L)
ax[1][1].set_xlabel('Eixo X')
ax[1][1].set_ylabel('Eixo L')
plt.tight_layout()
```

**Matplotlib - Gráfico de torta**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ["A1","A2","A3","A4","A5"]
```

```python
valores = [10,45,32,56,23]
```

```python
#Criando gráfico de barras
plt.pie(valores,labels=labels)
plt.show()
```

```python
#Inserindo porcentagem
plt.pie(valores,labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#Explode
explode = [0,0.1,0,0,0]
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%',explode=explode)
plt.show()
```
