#Matplotlib - Legenda

```python
#Importando Bibliotecas
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Criando variável X
X = np.linspace(0,5,11)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
# Colocando equações na legenda
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=E^{(X/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
#Adicionando posição da legenda 
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend(loc='lower right')
plt.show()
```

```python
#Criando novas variáveis X, Y e Z 

X = np.linspace(0,2.0*np.pi,361)
Y = np.cos(X/2)
Z = np.sin(X/4)
```

```python
plt.plot(X,Y,label='$\cos(X/2)$',color='red')
plt.plot(X,Z,label='$\sin(X/4)$',color='green',linestyle='--')
plt.legend()
```
