```python
#Importando as Bibliotecas de classes do projeto.
import numpy as np # Trabalha com objetos vetor e matriz multidimensional
import matplotlib.pyplot as plt #Gerar os gráficos de visualização.   
```

```python
#Criando os campos descritivos (também chamados de labels) (Rótulos de dados)
labels = ["A1", "A2", "A3", "A4", "A5"]
```

```python
valores = [10, 34, 56, 78, 12]
```

```python
#Visualilzando o gráfico oui objeto plt (MAtplotLIb)
plt.bar(labels, valores)
```

```python
#Visualilzando o gráfico na horizontal o objeto plt (MAtplotLIb)
plt.barh(labels, valores)
```

```python
#Atribuindo nomes aos eixos do gráfico (x,y)
plt.bar(labels,valores)
plt.xlabel('Código do produto')
plt.ylabel('Valor do produto')
```

```python
#Alterar as cores dos elementos gráficos
plt.bar(labels,valores, color='silver')
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
#Colocando valores existenstes em cada tipo como rótulo de sdados
plt.bar(labels,valores, color='red')
plt.xlabel('Código do Produto')
plt.ylabel('Valores dos produtos')

for x,y in zip (labels, valores):
  # plote em texto x e y com espaçamento entre eles de 0.05  %d %y centralizado va bottom - em cima do gráfico vc up -ficaria embaixo
    plt.text(x, y+0.05, '%d' % y, ha='center', va='bottom')
```

```python
#Alterar as cores dos elementos gráficos
plt.pier(labels,valores)
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
#Criando um novo vetor chamdo valores2.
valores2 = [8, 30, 60, 21, 30]
```

```python
largura = 0.3
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1, valores, color='red', width=largura, label='Barra 1')
plt.bar(pos2, valores2, color='blue', width=largura, label='Barra 2')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1")
plt.bar(pos1,valores2,label = "Barra 2",bottom = valores)
```
